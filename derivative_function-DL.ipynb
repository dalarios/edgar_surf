```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**2 + 2*x + 1

# Define the derivative of the function
def f_prime(x):
    return 2*x + 2

# Generate x values
x = np.linspace(-10, 10, 400)

# Compute y values for the function
y = f(x)

# Compute y values for the derivative
y_prime = f_prime(x)

# Plot the function
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='f(x) = x^2 + 2x + 1')
plt.plot(x, y_prime, label="f'(x) = 2x + 2", linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function and its Derivative')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Define the symbolic variable
x = sp.symbols('x')

# Define the function
f = x**3 + 2*x**2 + 4*x + 2

# Compute the derivative
f_prime = sp.diff(f, x)

# Convert the symbolic functions to numerical functions
f_num = sp.lambdify(x, f, 'numpy')
f_prime_num = sp.lambdify(x, f_prime, 'numpy')

# Generate x values
x_vals = np.linspace(-10, 10, 400)

# Compute y values for the function and its derivative
y_vals = f_num(x_vals)
y_prime_vals = f_prime_num(x_vals)

# Plot the function and its derivative
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label=f'f(x) = {sp.simplify(f)}')
plt.plot(x_vals, y_prime_vals, label=f"f'(x) = {sp.simplify(f_prime)}", linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function and its Derivative')
plt.legend()
plt.grid(True)
plt.show()
```
